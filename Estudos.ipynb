#Desafio 28 JOGO: Escreva um programa que faça o computador 'pensar' em um numero inteiro entre 0 e 5 e peça para o usuario tentar descobrir qual foi o numero escolhido pelo computador. O programa deverá escrever na tela se o usuario venceu ou perdeu.

```python
from random import randint  # Importa a função randint da biblioteca random para gerar números aleatórios
from time import sleep  # Importa a função sleep da biblioteca time para fazer o programa "esperar"

computador = randint(0, 5)  # Faz o computador sortear aleatoriamente um número entre 0 e 5

print('-=-' * 20)  # Imprime uma linha de separação estilizada
print('Vou pensar em um número entre 0 e 5. Tente adivinhar...')  # Mensagem ao jogador
print('-=-' * 20)  # Outra linha de separação

jogador = int(input('Em que número eu pensei? '))  # Pede para o jogador digitar um número e converte para inteiro

print('\033[1;31mPROCESSANDO...\033[m')  # Imprime a palavra "PROCESSANDO..." em vermelho (usando código ANSI para cor)

sleep(2)  # Faz o programa "dormir" por 2 segundos antes de mostrar o resultado, simulando processamento

if jogador == computador:  # Verifica se o número escolhido pelo jogador é igual ao número sorteado
    print('PARABÉNS! Você conseguiu me vencer!')  # Mensagem de vitória
else:
    print('VOCÊ PERDEU! Eu pensei no número {} e não no {}!'.format(computador, jogador))  # Mensagem de derrota, mostrando os dois números
```

```python
expre = str(input('Digite a expressão: '))
pilha = []
for simbolo in expre:
  if simbolo == '(':
    pilha.append('(')
  elif simbolo == ')':
    if len(pilha)> 0:
      pilha.pop()
    else:
      pilha.append(')')
      break
if len(pilha) == 0:
  print('Sua expressão esta valida!')
else:
  print('Sua expressão esta errada!')

```

#**Desafio 88 Faça um programa que ajude um jogador da MEGA SENA a criar palpites. O programa vai perguntar quantos jogos serão gerados e vai sortear 6 números entre 1 e 60 para cada jogo, cadastrando tudo em uma lista composta.**

```python
from random import randint
from time import sleep

print('-' * 40)
print(f'{"JOGA NA MEGA SENA":^40}')
print('-' * 40)

# Pergunta quantos jogos o usuário quer
quantidade = int(input('Quantos jogos você quer que eu sorteie? '))

# Lista que vai guardar todos os jogos
jogos = []

# Gera cada jogo
for j in range(quantidade):
    jogo = []
    while len(jogo) < 6:
        numero = randint(1, 60)
        if numero not in jogo:  # Evita repetir números
            jogo.append(numero)
    jogo.sort()  # Deixa os números em ordem crescente
    jogos.append(jogo)

# Mostra os jogos gerados
print('-' * 40)
print(f'{"SORTEANDO OS JOGOS":^40}')
print('-' * 40)

for i, jogo in enumerate(jogos, start=1):
    print(f'Jogo {i}: {jogo}')
    sleep(1)  # Faz uma pausa de 1 segundo entre os jogos

print('-' * 40)
print(f'{"BOA SORTE!":^40}')
print('-' * 40)
```

Crie uma lista principal (galera) e um dicionário temporário (pessoa).

Percorra a lista principal para encontrar as mulheres e as pessoas com idade acima da média.

```python
galera = []   # lista principal
pessoa = {}   # dicionário temporário
soma_idade = 0

while True:
    pessoa['nome'] = input("Nome: ")
    pessoa['sexo'] = input("Sexo [M/F]: ").strip().upper()
    pessoa['idade'] = int(input("Idade: "))
    soma_idade += pessoa['idade']
    galera.append(pessoa.copy())  # adiciona uma cópia do dicionário

    opcao = input("Quer continuar? [S/N] ").strip().upper()
    if opcao == 'N':
        break

# A) Quantas pessoas foram cadastradas
print(f"A) Total de pessoas cadastradas: {len(galera)}")

# B) Média de idade
media = soma_idade / len(galera)
print(f"B) Média de idade do grupo: {media:.2f}")

# C) Lista de mulheres
mulheres = [p['nome'] for p in galera if p['sexo'] == 'F']
print(f"C) Mulheres cadastradas: {mulheres}")

# D) Pessoas com idade acima da média
acima_media = [p['nome'] for p in galera if p['idade'] > media]
print(f"D) Pessoas com idade acima da média: {acima_media}")
```

Desafio 98: Função de Contador
Enunciado: Faça um programa que tenha uma função chamada contador(), que receba três parâmetros: início, fim e passo. Realize três contagens através da função:
A) De 1 até 10, de 1 em 1.B) De 10 até 0, de 2 em 2.C) Uma contagem personalizada que o usuário vai digitar.Conceitos-chave: Definição de funções com 3 parâmetros. Uso de laço (for) para a contagem. Lógica de contagem crescente e decrescente, incluindo a validação do passo (se for zero, considere 1).Proposta de Solução (Estrutura):Defina a função contador(i, f, p) que trata a contagem:Se o passo for $0$, mude-o para $1$.Se o início for menor que o fim (contagem crescente), use range(i, f + 1, p).Se o início for maior que o fim (contagem decrescente), use range(i, f - 1, -abs(p)) (o passo sempre deve ser negativo).Exiba a contagem pausadamente (ex: sleep(0.5) se importar o módulo time).Chame a função para os casos A e B.Peça ao usuário o início, fim e passo da contagem personalizada.Chame a função com os dados do usuário (caso C).

```python
from time import sleep

def contador(i, f, p):
    if p == 0:   # passo não pode ser zero
        p = 1
    print(f"Contagem de {i} até {f} de {p} em {p}:")

    if i < f:  # contagem crescente
        for c in range(i, f+1, p):
            print(c, end=' ', flush=True)
            sleep(0.5)
    else:      # contagem decrescente
        for c in range(i, f-1, -abs(p)):
            print(c, end=' ', flush=True)
            sleep(0.5)
    print("FIM!")

# Programa principal
print("-="*20)
contador(1, 10, 1)   # A) de 1 até 10, de 1 em 1
print("-="*20)
contador(10, 0, 2)   # B) de 10 até 0, de 2 em 2
print("-="*20)

print("Agora é sua vez de personalizar a contagem!")
inicio = int(input("Início: "))
fim = int(input("Fim: "))
passo = int(input("Passo: "))
contador(inicio, fim, passo)   # C) personalizada
```
